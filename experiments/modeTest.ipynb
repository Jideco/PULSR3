```python
# Open the input file
with open("realBigData.txt", "r") as file:
    lines = file.readlines()

# Initialize lists for upper and lower targets
upper_targets = []
lower_targets = []

# Loop through each line
for line in lines:
    # Split the line into words
    words = line.split()
    
    # Check if the line contains target values
    if "Upper Target" in line and "Lower Target" in line:
        # Extract and append the upper target value
        upper_target_index = words.index("Target") + 1
        upper_targets.append(int(words[upper_target_index]))
        
        # Extract and append the lower target value
        lower_target_index = words.index("Lower") + 2
        lower_targets.append(int(words[lower_target_index]))

# Write the upper targets to a file
with open("Bupper_targets.txt", "w") as file:
    for target in upper_targets:
        file.write(str(target) + "\n")

# Write the lower targets to a file
with open("Blower_targets.txt", "w") as file:
    for target in lower_targets:
        file.write(str(target) + "\n")
```
